```python
# 레스토랑 팁 데이터 셋 활용

'''
미션 : tip을 어떤 사람이 많이 주는지?
분석 - 데이터 취합

'''

tips = sns.load_dataset("tips")
```

```python
# total_bill 즉 계산한 금액과 tip 관계도 확인 - 산점도
# 축별 tip? total_bill? 라벨링
# 부산스러운 주소값 안나오게 하려면 세미콜론(;)을 뒤에 찍으면 됨.

sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용
# style must be one of white, dark, whitegrid, darkgrid, ticks

sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용, feature 를 색으로 구분
# 흡연 비흡연에 대한 색을 구분 - 범례 표현도 자동으로 설정

sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker');
```

```python
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='smoker');
```
