## 데이터 읽기

```python
cols = ['최근2년90일이상연체여부', '유동가능자산비율', '나이', '최근2년30-59일연체횟수', '총지출(부채비율)', '월수입', '신용대출 건수', '90일이상연체횟수', '부동산담보대출건수', '최근2년60-89일연체횟수', '본인재외부양가족수']
```

```python
import numpy as np
import pandas as pd

data = pd.read_csv("data/cs_data.csv")
```

```python
data = data.drop(columns=data.columns[0])
```

```python
data.columns = cols
data.head()
```

```python
data = pd.DataFrame(data)
data
```

## 이상치 처리

```python
# ['최근2년90일이상연체여부'] : 이상치 처리 불필요
a = df[['최근2년90일이상연체여부']].sort_values(by=['최근2년90일이상연체여부'], ascending=False)
a
```

```python
# ['유동가능자산비율'] : 이상치 처리 필요
b = df[['유동가능자산비율']].sort_values(by=['유동가능자산비율'], ascending=False)
b
```

```python
df[['유동가능자산비율']].mean()
```

```python
# ['나이'] : 이상치 처리?
c = df[['나이']].sort_values(by=['나이'], ascending=False)
c
```

```python
# ['최근2년30-59일연체횟수'] : 이상치 처리?
d = df[['최근2년30-59일연체횟수']].sort_values(by=['최근2년30-59일연체횟수'], ascending=False)
d
```

```python
# ['총지출(부채비율)'] : 이상치 처리 필요? 평균?
e = df[['총지출(부채비율)']].sort_values(by=['총지출(부채비율)'], ascending=False)
e
```

```python
# ['월수입'] : 평균???
f = df[['월수입']].sort_values(by=['월수입'], ascending=False)
f
```

```python
# ['신용대출 건수'] : 어쩔까?
g = df[['신용대출 건수']].sort_values(by=['신용대출 건수'], ascending=False)
g
```

```python
# ['90일이상연체횟수'] : 이상치 처리 불필요
h = df[['90일이상연체횟수']].sort_values(by=['90일이상연체횟수'], ascending=False)
h
```
