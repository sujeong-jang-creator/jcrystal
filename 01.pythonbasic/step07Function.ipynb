### 2. 키워드 인수

- **kwargs : keyword argument 줄임말
- **kwargs : 키워드=특정값 형태로호출

```python
def fun2(x, y, z):
    return x + y + z    
```

```python
fun2(10, 20, 30)
```

```python
# **kwargs 사용
def fun3(**kwargs):         # x=10, y=20, z=30
    print(type(kwargs))      # <class 'dict'>
    print(kwargs)            # {'x': 10, 'y': 20, 'z': 30}
    print(kwargs['x'])       # dict 데이터 사용 문법으로 활용
    print(kwargs.items())    # [('x', 10), ('y', 20), ('z', 30)]
```

```python
fun3(x=10, y=20, z=30)
```

```python
# **kwargs 사용 + 반복문과 items()
def fun4(**kwargs):
    for v in [('x', 10), ('y', 20), ('z', 30)]:
        print(v)

```

```python
fun4(x=10, y=20, z=30)
```

```python
def fun4(**kwargs):
    for k,v in [('x', 10), ('y', 20), ('z', 30)]:
        print(k, v)
```

```python
fun4(x=10, y=20, z=30)
```

```python
def fun4(**kwargs):
    for k,v in kwargs.items():
        print(k, ':' ,v)
```

```python
fun4(x=10, y=20, z=30)
```

```python
import random

def genPass():
    # 제공되는 raw 데이터
    alphabet = "abcdefghijklmnopqrstuvwxyz0123456789"
    
    # 6개 착출된 데이터를 보유하게 되는 변수
    # 반복을 6번 해서 6개의 문자열 구성
    # 데이터가 적재되는 변수
    password = ""
    
    for i in range(6):# 0~5까지 6개의 숫자 반환, 6번으로 사용되는 반복문
        index = random.randrange(36)
        password = password + alphabet[index]    # password += alphabet[index]
        
    return password


print(genPass())
print(genPass())
print(genPass())
```

```python
# key는 딱 정해져 있는 거래요.
# 매개변수가 a 가 되고 a[2]라는건 100, 99, 102 얘네들!
# list 내에 존재하는 각각의 tuple들의 index 2 즉 세번째 요소(데이터) 들로 오름차순 정렬
mylist2.sort(key=lambda a : a[2])
mylist2
```

```python
# 대소문자 구분없이 sort 함수로 정렬 
# 람다 사용해서 소문자로 변경해서 정렬 
# lower() - 소문자로 변경하는 함수
# upper() - 대문자로 변경하는 함수
mylist3 = ['Spam', 'egg', 'Ham']

for v in mylist3:
    print(v.lower())   # 대소문자의 유니코드 또는 아스키 코드의 번호를 고려한 대소문자 맞춤
    
    
print(mylist3)

# 원본 데이터 정렬만 관리
# 람다식으로 처리한 기능은 원본 데이터 활용해서 임시로
# 변경된 후에 휘발되는 구조
# key = lambda x : x.lower()

'''
['Spam', 'egg', 'Ham']
Spam, egg, Ham
S, e, H 뽑아서 아스키 또는 유니코드의 숫자 체계 비교
문제는 대소 문자의 숫자가 비교에 부적합
따라서 소문자로 다 변경 또는 대문자로 변경해서 비교하는 로직이 필요
로직은 대(소)문자로 변경하고자 하는게 주요 로직이 아니라
단순 정렬이 주요 로직이 중요한데
데이터 손실이 발생되는 것을 방지하고자 sort() 정렬 책임지고
lambda식은 대소문자 변경으로만 처리
Spam -> spam(메모리에 생성되어 존재하기는 함) / egg -> egg / Ham -> ham으로 비교
메모리에서 휘발(메모리에서 삭제의미)

'''


mylist3.sort(key = lambda x : x.lower())   # 비교할때 소문자 구분하지 말고 정렬해줘라 순서를!

print(mylist3)
```

### 5. map() 함수

- 문법 : map(함수, 리스트)
- 시퀀스의 모든 항목에 함수를 적용한 결과 리스트 반환

```python
def cube(x):
    return x**2

list(map(cube, range(1, 11)))    # 1, 11 - 1~10
```

```python
list(map(lambda x : x**2 ,range(1, 11)))
```

```python
for i in range(5):
    print(i**2)
```

```python
# map() : 다수의 데이터에 공통된 로직 일괄 처리 가능한 기능의 함수
# 사용 방법 : map(기능보유한함수명 or 람다식
#                 첫번째 parameter로 입력된 기능이 사용할 데이터)
# 대량의 빅데이터 일괄처리 하는 기능의 sw 대표적인 하둡 내부에서도 사용되는 기능

def myfunc(n):
    print(n)
    return len(n)

# tuple
x = map(myfunc, ('apple', 'banana', 'cherry'))

print(x)    # <map object at 0x0000018F1D787370
print(list(x))
```

```python
def myfunc(n):
    print(n)
    return len(n)

# list
x = map(myfunc, ('apple', 'banana', 'cherry'))


print(list(x))
```

### 6. reduce() 함수
- 주로 집계용으로 사용
- 사용을 위해서는 functools 모듈 선언

```python
# 펑션이라는 폴더 하위에 리듀스가 있다는 식으로 이해
# 무조건 전의 결과값이 x로 됨.

from functools import reduce

'''

x, y
    1, 2 => 3
    3, 3 => 6
    6, 4 => 10
    10, 5 => 15

'''

data = reduce(lambda x, y: x+y, [1, 2, 3, 4, 5])
print(data)
```

```python
#상단 reduce에 사용된 lambda의 로직을 단순 python 문법으로 개발한 코드

def userReduce():
    datas = [1, 2, 3, 4, 5]
    
    r = 0
    
    for x in datas:
        r = r + x
        
    print(r)

userReduce()
```
