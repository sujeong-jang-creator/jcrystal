**특수 문자 판단하기**

- 정규표현식에 사용하는 특수 문자 *, +, ?, ., ^, $, (, ) [, ], - 등을 판단하려면 특수 문자를 판단할 때는 특수 문자 앞에 \ 표현
- \d : [0-9] 표현과 동일
- \D : [^0-9] 표현과 동일, 숫자를 제외한 모든 문자
- \w : [a-zA-Z0-9_] 표현과 동일, 영문 대소문자, 숫자, 밑줄문자
- \W : [^a-zA-Z0-9_] 표현과 동일, 영문 대소문자, 숫자, 밑줄문자를 제외한 모든 문자

**데이터 그룹핑 하기**
- 그룹핑이란? 특정 데이터가 계속해서 반복하는지를 조사할때 사용 가능한 정규식<br>
- grouping 표기 : ()


**그룹핑 관련 함수**
- group() or group(0) : 매치된 전체 문자열
- group(1) : 첫번째 그룹에 해당하는 문자열 반환
- group(n) : n번째 그룹에 해당하는 문자열 반환


```python
import re
```

```python
m = re.match('([0-9]+) ([0-9]+)', '150 22')
print(m)
```

```python
# 해당 문자열을 한번에 반환

print(m.group(0))
```

```python
# 첫 번째 그룹에 매핑된 문자열 반환

print(m.group(1))
```

```python
# 두 번째 그룹에 매핑된 문자열 반환

print(m.group(2))
```

"이름 000-0000-0000"과 같은 전화번호 형태의 
정규 표현식 만들고 데이터 그룹 이해하기

- 그룹 개수가 많아지면 숫자로 그룹을 구분하기가 쉽지 않음
- 그룹에 이름을 적용하여 편리하게 구분
- syntax : (?P<이름>정규표현식)

```python
# 함수 이름과 인수값 추출
import re
m = re.match('(?P<funcName>[a-zA-Z][a-zA-Z0-9]+)\((?P<arg>\w+)\)', 'print(1234)' )

# 그룹 이름으로 매칭된 문자열 출력
m.group('funcName')
```

```python
# 그룹 이름으로 매칭된 문자열 출력
m.group('arg')
```

### 전후방 탐색(lookaround)

- 의미 : 원하는 문자열 검색 범위 지정시 사용
- **전방탐색(lookahead) 패턴** : 앞에서 찾기, 일치 영역 발견 후 해당 문자열을 제외한 앞부분 반환
- 전방탐색 syntax : ?=
<br>

- **후방탐색(lookbehind) 패턴** : 뒤에서 찾기, 제시된 패턴의 뒷쪽을 탐색 및 반환
- 후방탐색 syntax : ?<=
