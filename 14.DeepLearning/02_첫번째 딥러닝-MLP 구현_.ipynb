```python
# ()
model = keras.Sequential()
#  Layer 
model.add(keras.layers.InputLayer((2,2)))  # Layer Input Layer ( shape)
model.add(keras.layers.Flatten()) # 28 x 28  1 (784,) 
model.add(keras.layers.Dense(units=128, activation='relu'))
model.add(keras.layers.Dense(units=64, activation='relu'))
model.add(keras.layers.Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

##  
-      
- 
- Optimizer( )
- 

```python
model.compile(optimizer="adam", #- (weight) 
              loss="categorical_crossentropy", # Loss(,)  . categorical_crossentropy:  
              metrics=['accuracy'] #loss   
             )
```

##  
- X (Input Data Image)
    - 0 ~ 1    . (Scaling)
- y (Output Data)
    - one hot encoding 
        - Label   One Hot Encoding .
    - tensorflow.keras  to_categorical()  

```python
np.min(train_image), np.max(train_image), train_image.dtype
```

```python
X_train = (train_image/255.0).astype(np.float32)
```

```python
X_test = (test_image/255.0).astype(np.float32)
```

```python
np.min(X_train), np.max(X_train), X_train.dtype
```

```python
train_label, np.unique(train_label)
```

```python
train_label[:5]
```

```python
# label onehot encoding
y_train = keras.utils.to_categorical(train_label)
y_test = keras.utils.to_categorical(test_label)
```

```python
print(y_train.shape, train_label.shape)
y_train[:5]
```

##  (fit)

```python
# model.fit() : History  
# History : train     .
```

```python
# 1 epoch: train dataset    .
# 1 step : (weight)  update
hist = model.fit(X_train, y_train, # train dataset - x, y
                  epochs=10,  #epoch  - train dataset   .
                  batch_size=100 , #(weight- )    . - 
                  validation_split=0.3 #train set 30%   .
                 )
#  : 60000, batch_size: 100
# 1 epoch   600 step
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,10))

plt.subplot(2,1,1)
plt.plot(range(1,11), hist.history['loss'], label='Train Loss')
plt.plot(range(1,11), hist.history['val_loss'], label="Validation Loss")
plt.title("Loss per Epoch", fontsize=20)
plt.legend()

plt.subplot(2,1,2)
plt.plot(range(1,11), hist.history['accuracy'], label='Train Accuracy')
plt.plot(range(1,11), hist.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy per Epoch", fontsize=20)

plt.legend()
plt.tight_layout()
plt.show()
```

##  

```python
test_loss, test_acc = model.evaluate(X_test, y_test)  #(X, y). : loss [, metrics ]
```
